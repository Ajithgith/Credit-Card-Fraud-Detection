```python
X = credit_data.drop(['Class'], axis=1)
Y = credit_data['Class']
```

```python
X
```

```python
def find_outliers(data):
  for col in data.columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3-Q1

    upper_limit = Q3+1.5*IQR
    lower_limit = Q1-1.5*IQR

    count_outliers_upper_limit = data[data[col]>upper_limit].shape[0]
    count_outliers_lower_limit = data[data[col]<lower_limit].shape[0]

    total_outliers = count_outliers_upper_limit+count_outliers_lower_limit

    print(f'{col}:\n')

    print("Upper Limit:", upper_limit)
    print("Lower Limit:", lower_limit)
    print("Upper Limit Outliers:", count_outliers_upper_limit)
    print("Lower Limit Outliers:", count_outliers_lower_limit)
    print(f'{total_outliers} outliers out of {data.shape[0]}. i.e.; {total_outliers/data.shape[0]*100} % of outliers\n\n')
```

```python
find_outliers(X)
```

```python
def distribution_of_columns(col):
    print(f'{col}:\n')
    sns.distplot(X[col])
    print('\n')
```

```python
def replace_outliers(data):
  for col in data.columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3-Q1

    upper_limit = Q3+1.5*IQR
    lower_limit = Q1-1.5*IQR

    data.loc[data[col]<lower_limit, col] = lower_limit
    data.loc[data[col]>upper_limit, col] = upper_limit

  return data
```

```python
replace_outliers(X)
```

```python
find_outliers(X)
```

Handled the outliers in all the numerical features.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
10.0 **np.arange(-2,3)
```

```python
log_class = LogisticRegression()
grid = {'C': 10.0 **np.arange(-2,3), 'penalty':['l1', 'l2']}
cv = KFold(n_splits=5, shuffle=False, random_state=None)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
clf = GridSearchCV(log_class, grid, cv=cv, n_jobs=-1, scoring='f1_macro')
clf.fit(X_train, Y_train)
```

```python
Y_pred = clf.predict(X_test)
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=10, criterion='gini', random_state=None)
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler

os = RandomOverSampler()
X_train_os, Y_train_os = os.fit_resample(X_train, Y_train)
print("The number of classes before fit {}".format(Counter(Y_train)))
print("The number of classes after fit {}".format(Counter(Y_train_os)))
```

```python
cla = RandomForestClassifier(n_estimators=10, criterion='gini', random_state=None)
cla.fit(X_train_os, Y_train_os)
```

```python
Y_pred = cla.predict(X_test)
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
from imblearn.combine import SMOTETomek

sm = SMOTETomek()
X_train_sm, Y_train_sm = sm.fit_resample(X_train, Y_train)
print("The number of classes before fit {}".format(Counter(Y_train)))
print("The number of classes after fit {}".format(Counter(Y_train_sm)))
```

```python
clp = RandomForestClassifier(n_estimators=10, criterion='gini', random_state=None)
clp.fit(X_train_sm, Y_train_sm)
```

```python
Y_pred = clp.predict(X_test)
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```
